```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv("data/mouse_drug_data.csv")
clinical_trial_df = pd.read_csv("data/clinicaltrial_data.csv")

# Combine the data into a single dataset
combined_pymaceuticals_data = pd.merge(clinical_trial_df, mouse_drug_df, on="Mouse ID")

# Display the data table for preview
combined_pymaceuticals_data.sort_values(by='Timepoint', ascending=True).head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame
```

```python
# Retrieve Drug, Timepoint,Tumor Volume (mm3) column data
tumor_data = combined_pymaceuticals_data[["Drug", "Timepoint","Tumor Volume (mm3)"]]
# Grouped by Drug and Timepoint
avg_tumor_vol= tumor_data.groupby(["Drug","Timepoint"])
# Mean tumor volume (mm3)
tumor_response = avg_tumor_vol.mean()
tumor_response = tumor_response.reset_index()
tumor_response
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame

```

```python
# Standard Error of Tumor Volumes Grouped by Drug and Timepoint

std_err_tumor = avg_tumor_vol.sem()
std_err_tumor =std_err_tumor.reset_index()
std_err_tumor.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
```

```python
reformat_tumor_resp = tumor_response.reset_index().pivot("Timepoint", "Drug", "Tumor Volume (mm3)")
reformat_tumor_resp.head()
```

```python
# Generate the plot (with Error Bars)
#Save the Figure
```

```python
reformat_std_err_tumor = std_err_tumor.reset_index().pivot("Timepoint", "Drug", "Tumor Volume (mm3)")
reformat_std_err_tumor.head()
```

```python
# Generate the Plot (with Error Bars) 
plt.errorbar(reformat_tumor_resp.index,reformat_tumor_resp["Capomulin"],yerr= reformat_std_err_tumor["Capomulin"],color="r",
             marker="o", markersize= 5,linestyle = "dashed", linewidth = 0.5, label= "Capomulin")
plt.errorbar(reformat_tumor_resp.index,reformat_tumor_resp["Infubinol"],yerr= reformat_std_err_tumor["Infubinol"],color="b",
             marker="^", markersize= 5,linestyle = "dashed", linewidth = 0.5, label = "Infubinol" )
plt.errorbar(reformat_tumor_resp.index,reformat_tumor_resp["Ketapril"],yerr= reformat_std_err_tumor["Ketapril"],color="g",
             marker="s", markersize= 5,linestyle = "dashed", linewidth = 0.5, label = "Ketapril")
plt.errorbar(reformat_tumor_resp.index,reformat_tumor_resp["Placebo"],yerr= reformat_std_err_tumor["Placebo"],color="k",
             marker="d", markersize= 5,linestyle = "dashed", linewidth = 0.5, label = "Placebo")

# Save the Figure
plt.title("Tumor Response to Treatment")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Time (Days)")
plt.grid(True)
plt.legend(loc="best",fontsize ="small",fancybox=True)
#plt.show()
plt.savefig("treatment_new.png")
plt.show()
```

```python
#Survival Rate 
mouse_count=combined_pymaceuticals_data.groupby(["Drug","Timepoint"]).nunique()["Mouse ID"]
#Compute Survival Rate 
mouse_survival_rate = mouse_count.copy()

current_drug = "" 
current_drug_max_mouse_count = 0
for key, value in mouse_survival_rate.items() :
    d, t = key
    if t == 0:
        current_drug_max_mouse_count = value
    mouse_survival_rate[key] = (float(value)*100)/float(current_drug_max_mouse_count)           


# Display the result in a dataframe
survival_rate =pd.DataFrame({"Mouse Count":mouse_count, "Mouse Survival Rate": mouse_survival_rate})
survival_rate =survival_rate.reset_index()
survival_rate.head()
```

```python
reformat_survival_rate = survival_rate.reset_index().pivot("Timepoint", "Drug","Mouse Count")
reformat_survival_rate = survival_rate.reset_index().pivot("Timepoint", "Drug", "Mouse Survival Rate")
reformat_survival_rate.head()
```

```python
# Generate the Plot (with Error Bars) 
plt.errorbar(reformat_survival_rate.index,reformat_survival_rate["Capomulin"],yerr= 0,color="r",
             marker="o", markersize= 5,linestyle = "dashed", linewidth = 0.5, label = "Capomulin")
plt.errorbar(reformat_survival_rate.index,reformat_survival_rate["Infubinol"],yerr= 0,color="b",
             marker="^", markersize= 5,linestyle = "dashed", linewidth = 0.5,label = "Infubinol")
plt.errorbar(reformat_survival_rate.index,reformat_survival_rate["Ketapril"],yerr= 0,color="g",
             marker="s", markersize= 5,linestyle = "dashed", linewidth = 0.5, label= "Ketapril")
plt.errorbar(reformat_survival_rate.index,reformat_survival_rate["Placebo"],yerr= 0,color="k",
             marker="d", markersize= 5,linestyle = "dashed", linewidth = 0.5, label = "Placebo")

# Save the Figure
plt.title("Survival During Treatment")
plt.ylabel("Survival Rate(%)")
plt.xlabel("Time(days)")
plt.grid(True)
plt.legend(loc="best",fontsize ="small",fancybox=True)
#plt.show()
plt.savefig("survival_new.png")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug

# Display the data to confirm
```

```python
tumor_pct_change = dict()
reformat = combined_pymaceuticals_data.groupby(["Drug"])
#reformat.head()
reformat.groups.keys()
for k in reformat.groups.keys() :
    tumor_vol_mean = reformat.get_group(k).groupby("Timepoint")["Tumor Volume (mm3)"].mean()
    tumor_vol_mean_list = tumor_vol_mean.tolist()
    first_item = tumor_vol_mean_list[0]
    last_item = tumor_vol_mean_list[len(tumor_vol_mean_list) - 1]
    tumor_vol_pct_change = ((last_item - first_item)/first_item)*100
    #print(k + "      " + str(tumor_vol_pct_change))
    tumor_pct_change[k] = tumor_vol_pct_change 
    print(k + "      " + str('{0:.6f}'.format(tumor_vol_pct_change)))
   
```

```python
# Store all Relevant Percent Changes into a Tuple


# Splice the data between passing and failing drugs


# Orient widths. Add labels, tick marks, etc. 


# Use functions to label the percentages of changes


# Call functions to implement the function calls


# Save the Figure


# Show the Figure
```

```python
#Store all Relevant Percent Changes into a Tuple
pct_changes = (tumor_pct_change["Capomulin"], 
               tumor_pct_change["Infubinol"], 
               tumor_pct_change["Ketapril"], 
               tumor_pct_change["Placebo"])
# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_changes))  
width = 1
rectsPass = ax.bar(ind[0], pct_changes[0], width, color='green')
rectsFail = ax.bar(ind[1:], pct_changes[1:], width, color='red')
# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)
# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")
def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")
# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)
# Save the Figure
fig.savefig("change_new.png")
# Show the Figure
fig.show()




```
